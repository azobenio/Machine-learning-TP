```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
#plot(x,y)==> données sur l'exe x et y
x=np.linspace(0,2,10)# tableau de 10 point allant de 0 à 2
y=x**2
#plt.plot(x,y)
```

```python
# premiere erreur à eviter 
# deux tableaux de dimensien differentes 
x=np.linspace(0,2,10)
y=np.linspace(0,10,10)
plt.plot(x,y)
```

```python
#plot(x,y)==> données sur l'exe x et y
x=np.linspace(0,2,10)# tableau de 10 point allant de 0 à 2
y=x**2
plt.scatter(x,y)
```

```python
#plot(x,y,label=....,lw=.....,ls=....,c=....)==> nomn courbe, epaissaire du trait, style trait, couleur trait
x=np.linspace(0,2,10)# tableau de 10 point allant de 0 à 2
y=x**2
plt.plot(x,y,c='red',lw=5,ls='--')
```

```python
#cycle de vie d'une figure 
#plt.figure(figsize=(12,8))==> creation d'une nouvelle figure==> vide 
#plt.plot(x,y)
#plt.plot(x,x**3)
#plt.title('titre')
#plt.xlabel('axe x')
#plt.ylabel('axe x')
#plt.legend()==> pour differencier les courbes , il faut valoriser l'attribut label 
```

```python
#cycle de vie d'une figure 
plt.figure(figsize=(12,8))
plt.plot(x,y,label="Fonction1")
plt.plot(x,x**3,label="Fonction2")
plt.title('titre')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('figure1.png')
```

```python
#plusieurs figure  
plt.figure(figsize=(12,8))
plt.plot(x,y,label="Fonction1")
plt.plot(x,x**3,label="Fonction2")
plt.title('titre')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.savefig('figure1.png')


#cycle de vie d'une figure 
plt.figure(figsize=(12,8))
plt.plot(x,y,label="Fonction1")
```
