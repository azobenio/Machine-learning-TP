```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x=np.arange(1,11)
x
```

```python
y=2*x
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
plt.plot(x,y, color="g" ,linestyle=':')
plt.title("Line plot")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# Adding two lines in the same plot
```

```python
y2 = 3*x
```

```python
y2
```

```python
plt.plot(x,y,color="g",linewidth=2)
plt.plot(x,y2,color="r",linewidth=5)
plt.title("two lines")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri)
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri,hue="event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", data=fmri,hue="event", style="event")
plt.show()
```

```python
sns.lineplot(x="timepoint", y="signal", 
             hue="event", style="event",
             markers=True, data=fmri)
plt.show()
```
